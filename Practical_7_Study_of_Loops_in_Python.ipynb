```python
# While Loop
# Used when the number of iterations is not known.
Syntax
while condition:
  statement(s)
```

```python
# Print i as long as i is less than 6:
i=1
while i<=5:
  print(i)
  i +=1
```

```python
# Print numbers from 1 to N
i=1
N=int(input("Enter the value of N: "))
while i<N:
  print(i)
  i+=1
```

```python
# Factorial of a Number
i=1
N=int(input("Enter the value of N: "))
f=1
while i<= N:
  f=f*i
  i+=1
print("Factorial of", N, "is", f)
```

```python
# Factorial of a Number 2
n=int(input("Enter the value of n: "))
fact=1
while n>0:
  fact*=n
  n-=1
print("Factorial=",fact)
```

```python
# Fibonacci Series using While Loop

a,b=0,1
count=0
n=int(input("Enter the number of terms: "))
print("THE SERIES IS :")
while count<n:
    print(a)
    a, b = b, a + b
    count += 1
```

```python
# Reverse a Number
n=int(input("Enter the number: "))
rev=0
while n>0:
    digit=n%10
    rev=(rev*10)+digit
    n=n//10
print("The reversed number is:", rev)

```

```python
# Check Palindrome Number
n=int(input("Enter the number: "))
m=n
rev=0
while n>0:
    digit=n%10
    rev= (rev*10)+digit
    n=n//10
if m==rev:
    print(m,"is a palindrome number")
else:
    print(m,"is not a palindrome number")
```

```python
# Check Palindrome for String
n=input("Enter the string: ")
i,j=0,len(n)-1
while i<j:
    if n[i]!=n[j]:
        print("Not a Palindrome")
        break
    i+=1
    j-=1
else:
    print("Palindrome")
```

```python
# Count Digits in a Number
n=int(input("Enter the number: "))
count=0
while n>0:
    n=n//10
    count+=1
print(count)
```

# The Break Statement
## With the break statement we can stop the loop even if the while condition is true:

```python
# Exit the loop when i is 3:
i=1
while i<6:
  print(i)
  if i==3:
    break
  i+=1
```

```python
# Search an element in a list
nums=[10,20,30,40,50]
key=int(input("Enter the element to search: "))
i=0
while i<len(nums):
    if nums[i]==key:
      print("Element found at index",i)
      break
    i+=1
else:
  print("Element not Found")
```

```python
# Print numbers from 1 to 10 (skip 5)

i=1
while i<=10:
  if i == 5:
    i+=1
    continue
  print(i)
  i+=1
```

```python
# Print only odd numbers (1 to 10)
i=1
while i<=10:
  if i%2== 0:
    i+=1
    continue
  print(i,end=" ")
  i+=1
```

```python
# Multiplication of Two 3x3 Matrices (most important)
A=[[1,2,3],
 [4,5,6],]
B=[[1,1,1],
 [6,5,4],
  [3,2,1]]

Result=[[0,0,0],
 [0,0,0]]

for i in range(2):
  for j in range(3):
    for k in range(2):
      Result[i][j]+=A[i][k]*B[k][j]


print("Product of Matrices")
for row in Result:
  print(row)
```

```python
#Accept 3 digit number and print all possible combination
N=input('Enter 3 digit number:')
a=int(N[0])
b=int(N[1])
c=int(N[2])
d=[a,b,c]

for i in range(0,3):
  for j in range(0,3):
    for k in range(0,3):
      if (d[i]!=d[j] and d[j]!=d[k] and d[k]!=d[i]):
        print(d[i],d[j],d[k])
```

```python

#Armstrong Number
N=input('Enter a number: ')
total=0

for i in N:
  total+= int(i)**len(N)

if total==int(N):
  print(N," is an Armstrong Number")
else:
  print(N," is not an Armstrong Number")
```

```python
#Floyd's Series
n=int(input('Enter a number of rows of floyd series to be printed: '))
num=1
for i in range(1,n+1):
  for j in range(1,i+1):
    print(num, end=' ')
    num+=1
  print()
```
